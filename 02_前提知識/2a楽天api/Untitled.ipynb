```python
import requests
```

```python
REQUEST_URL = 'https://app.rakuten.co.jp/services/api/BooksBook/Search/20170404' 
APP_ID = "1034235205250283151"
```

```python
# param

params = {
    "format":"json",
    "formatVersion":1,
    "publisherName":"早川書房",
    "size":2,
    "page":1,
    "applicationId":APP_ID
}
```

```python
res = requests.get(REQUEST_URL, params)
res
```

```python
result = res.json()
```

```python
df = pd.DataFrame()

for i in range(0,len(result["Items"])):
    book_info = result['Items'][i]['Item'] 
    temp_df = pd.DataFrame(book_info,index=[i]) 
    df = pd.concat([df,temp_df]) 
    
display(df)
```
